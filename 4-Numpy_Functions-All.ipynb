```python
def arrayinfo(a):
    print(a)
    print('Data Type of a variable a is: ',type(a))
    print('Number of dimensions:',a.ndim)
    print('Shape of the nd array:',a.shape)
    print('Number of elements in the nd array:',a.size)
    print('ndarray element datatype:',a.dtype)
    print('Memory allogation for each element:',a.itemsize)
```

##### using an array is faster than list

```python
## Addition of 1000 numbers using list and numpy both the ways
```

```python
a = range(0,1000)
b = range(0,1000)
```

```python
import time
t1 = time.time()
for i in range(0,1000):
    d = a[i]+b[i]
    print(d)
t2 = time.time()
total_time = t2-t1
print(total_time)
```

```python
print(total_time)
```

```python
a = np.arange(0,1000)
b = np.arange(0,1000)
```

```python
import time
t1 = time.time()
for i in range(0,1000):
    d = a[i]+b[i]
    print(d)
t2 = time.time()
total_time = t2-t1
```
